## 游꿢 Objetivo general

El objetivo del proyecto es comprender los factores que llevan a la p칠rdida de clientes (*churn*) en la empresa ficticia **Telecom X**. A trav칠s del an치lisis exploratorio de datos (EDA), se buscar치n patrones y correlaciones para fundamentar decisiones de negocio y alimentar modelos predictivos.

## 游빌 Fase ETL - Extracci칩n

### 游댳 Acciones realizadas
- Identificamos como fuente de datos el archivo `TelecomX_Data.json`.
- Cargamos el archivo usando la librer칤a est치ndar `json`.
- Convertimos el contenido a una estructura Python (lista de diccionarios).
- Visualizamos un ejemplo para inspeccionar su estructura interna anidada.

> Esta etapa representa el inicio del proceso ETL: **Extracci칩n** de los datos desde su fuente original.


```python

import json
import pandas as pd

# Ruta del archivo JSON (ajustar si est치s en otro entorno)
ruta = '/content/TelecomX_Data.json'

# Cargar el archivo como lista de diccionarios
with open(ruta, 'r') as f:
    datos = json.load(f)

# Visualizar un registro para entender la estructura
print("Ejemplo de un registro:")
print(json.dumps(datos[0], indent=2))
```


# 游댢 Transformaci칩n

## 游꿢 Objetivo

Transformar los datos cargados desde el JSON, que tienen una estructura anidada (jer치rquica), en un `DataFrame` plano con columnas simples. Esto es esencial para aplicar an치lisis estad칤stico y visualizaci칩n.

## 游 Justificaci칩n

Las bibliotecas de an치lisis como `pandas` requieren estructuras tabulares. Por lo tanto, desanidar los campos es parte clave del subproceso de **Transformaci칩n** dentro del flujo ETL.

> Utilizaremos `pandas.json_normalize()` para aplanar la estructura, separando campos internos con un punto.


```python

# Aplicamos json_normalize para transformar a DataFrame plano
df = pd.json_normalize(datos, sep='.')

# Vista preliminar del DataFrame
df.head()
```

```python
# Ver forma general del DataFrame
print("Dimensiones del dataset (filas, columnas):", df.shape)

# Nombres de las columnas
print("\nColumnas disponibles:")
print(df.columns.tolist())

# Tipos de datos
print("\nTipos de datos:")
print(df.dtypes)

# O bien, un resumen m치s completo
print("\nResumen general del DataFrame:")
df.info()
```
