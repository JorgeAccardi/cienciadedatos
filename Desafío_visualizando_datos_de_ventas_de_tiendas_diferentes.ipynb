```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}
```

```python
df = pd.DataFrame(ventas_2022, index = tiendas)
```

```python
df
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,4))
plt.subplots_adjust(wspace=0.3, hspace=0.7)
fig.suptitle('Ventas en el período de Ene a Dic de 2022 en las tiendas A, B, C y D')
axs[0,0].plot(df.loc['A'])
axs[0,0].set_title('Ventas en la tienda A')
axs[0,1].plot(df.loc['B'])
axs[0,1].set_title('Ventas en la tienda B')
axs[1,0].plot(df.loc['C'])
axs[1,0].set_title('Ventas en la tienda C')
axs[1,1].plot(df.loc['D'])
axs[1,1].set_title('Ventas en la tienda D')

for ax in axs.flat:
  ax.set_xlabel('Mes')
  ax.set_ylabel('Número de Ventas')

plt.show()
```
