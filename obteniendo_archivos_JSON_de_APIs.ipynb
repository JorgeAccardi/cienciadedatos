##**Obteniendo archivos JSON de APIs**

```python
import pandas as pd
```

```python
import requests
```

```python
import json
```

```python
datos_frutas = requests.get('https://fruityvice.com/api/fruit/all')
```

```python
resultado = json.loads(datos_frutas.text)
```

```python
pd.DataFrame(resultado)
```
