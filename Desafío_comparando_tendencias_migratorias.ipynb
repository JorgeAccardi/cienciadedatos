## Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/inmigrantes_canada.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace = True)
```

```python
df
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
colombia = df.loc['Colombia', anos]
```

```python
colombia
```

```python
col_dict = {'Año': colombia.index.tolist(),
            'Inmigrantes': colombia.values.tolist()}
```

```python
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
```

```python
datos_col
```

```python
datos_col.tail()
```

## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (8,4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2014'])
plt.show()
```

```python
plt.figure(figsize = (8,4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2014'])
plt.title('Inmigración de Colombianos a Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.show()
```

```python
df_comparacion = df.loc[['Brasil','Argentina'], anos]
```

```python
df_comparacion
```

```python
df_comparacion = df_comparacion.T
```

```python
df_comparacion.head()
```

```python
plt.plot(df_comparacion['Brasil'], label='Brasil')
plt.plot(df_comparacion['Argentina'], label='Argentina')
plt.title('Inmigración de Brasil y Argentina a Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigración de Colombianos a Canadá\n 1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('Número Inmigrantes')
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize = (10,4))

axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'])
axs[0].set_title('Inmigrantes Colombianos a Canadá\n 1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Número de Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de Inmigrantes Colombianos a Canadá\n 1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Número de Inmigrantes')
axs[1].grid()

plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize = (10,6))
fig.subplots_adjust(wspace=0.5, hspace=0.3)
fig.suptitle('Inmigración de los 4 mayores paises sudamericanos a Canadá\n 1980 - 2013')

axs[0,0].plot(df.loc['Colombia', anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil', anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú', anos])
axs[1,1].set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

for axs in axs.flat:
  axs.xaxis.set_major_locator(plt.MultipleLocator(5))
  axs.set_xlabel('Año')
  axs.set_ylabel('Número de Inmigrantes')

plt.show()
```

```python
sudamerica = df.query('Region == "América del Sur"')
sudamerica.head(10)
```

```python
colores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown',
           'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize = (12,5))
ax.bar(sudamerica.index, sudamerica['Total'], color = colores)
ax.set_title('Inmigración de sudamerica a Canadá\n Período 1980 - 2013', loc='left', fontsize=16)
ax.set_xlabel('')
ax.set_ylabel('Número de inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
colores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown',
           'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize = (12,5))
ax.barh(sudamerica.index, sudamerica['Total'], color = colores)
ax.set_title('Inmigración de sudamerica a Canadá\n Período 1980 - 2013', loc='left', fontsize=16)
ax.set_ylabel('')
ax.set_xlabel('Número de inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
sudamerica_sorted = sudamerica.sort_values('Total', ascending=True)

colores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown',
           'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize = (12,5))
ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color = colores)
ax.set_title('Inmigración de sudamerica a Canadá\n Período 1980 - 2013', loc='left', fontsize=16)
ax.set_ylabel('')
ax.set_xlabel('Número de inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
colores = []
for pais in sudamerica_sorted.index:
  if pais == 'Argentina':
    colores.append('slateblue')
  else:
    colores.append('silver')

fig, ax = plt.subplots(figsize = (10,3))

ax.set_frame_on(False)
ax.xaxis.set_visible(False)
ax.tick_params(axis='y', which='both', size=0)

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color = colores)
ax.set_title('Sudamerica: Argentina es el Séptimo País con mas migrantes a Canadá\n Período 1980 - 2013', loc='left', fontsize=16)
ax.set_ylabel('')
ax.set_xlabel('Número de inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=12)


for i,j in enumerate(sudamerica_sorted['Total']):
  ax.text(j+20,i,str(j),color='black',fontsize=8.5,ha='left',va='center')

fig.savefig('/content/inmigrantes_sudamericanos_canada.png', transparent=False,
            dpi=300, bbox_inches='tight')
plt.show()

```
