```python
sudamerica_sorted = sudamerica.sort_values('Total', ascending=True)

colores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown',
           'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize = (12,5))
ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color = colores)
ax.set_title('Inmigración de sudamerica a Canadá\n Período 1980 - 2013', loc='left', fontsize=16)
ax.set_ylabel('')
ax.set_xlabel('Número de inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
colores = []
for pais in sudamerica_sorted.index:
  if pais == 'Argentina':
    colores.append('slateblue')
  else:
    colores.append('silver')

fig, ax = plt.subplots(figsize = (10,3))

ax.set_frame_on(False)
ax.xaxis.set_visible(False)
ax.tick_params(axis='y', which='both', size=0)

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color = colores)
ax.set_title('Sudamerica: Argentina es el Séptimo País con mas migrantes a Canadá\n Período 1980 - 2013', loc='left', fontsize=16)
ax.set_ylabel('')
ax.set_xlabel('Número de inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=12)


for i,j in enumerate(sudamerica_sorted['Total']):
  ax.text(j+20,i,str(j),color='black',fontsize=8.5,ha='left',va='center')

fig.savefig('/content/inmigrantes_sudamericanos_canada.png', transparent=False,
            dpi=300, bbox_inches='tight')
plt.show()

```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize = (8,4))
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax.tick_params(axis='y', which='both', size=0)

  ax = sns.barplot(data=top_10,x='Total', y=top_10.index, hue=top_10.index, orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 paises de mayor migración a Canadá\n Período 1980 a 2013',
              loc='left', fontsize=16)
  ax.set_xlabel('Número de Inmigrantes', fontsize=14)
  ax.set_ylabel('')

  for i,j in enumerate(top_10['Total']):
   ax.text(j+20,i,str(j),color='black',fontsize=8.5,ha='left',va='center')

  plt.show()
```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize = (8,4))
  ax.set(xticklabels = [])

  ax = sns.barplot(data=top_10,x='Total', y=top_10.index, hue=top_10.index, orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 paises de mayor migración a Canadá\n Período 1980 a 2013',
              loc='left', fontsize=16)
  ax.set_xlabel('Número de Inmigrantes', fontsize=14)
  ax.set_ylabel('')
  sns.despine()

  for i,j in enumerate(top_10['Total']):
   ax.text(j+20,i,str(j),color='black',fontsize=8.5,ha='left',va='center')

  plt.show()
```

```python
import plotly.graph_objs as go

# Criando uma figura
fig = go.Figure()

# Adicionando a linha do gráfico e definindo a espessura da linha
fig.add_trace(
    go.Scatter(x=[datos_col['Año'].iloc[0]], y=[datos_col['Inmigrantes'].iloc[0]], mode='lines', name='Inmigrantes', line=dict(width=4))
)

# Definir la configuración del layout
fig.update_layout(
    title=dict(
        text='<b>Inmigración de Colombianos hacia Canadá en el periodo de 1980 a 2013</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Año</b>'),
    yaxis=dict(range=[0, 7000], autorange=False, title='<b>Número de Inmigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1000,
    height=500
)

# Definir la configuración de la animación
frames = [go.Frame(data=[go.Scatter(x=datos_col['Año'].iloc[:i+1], y=datos_col['Inmigrantes'].iloc[:i+1])]) for i in range(len(datos_col))]
fig.frames = frames

# Mostrar la figura
fig.show()
```
