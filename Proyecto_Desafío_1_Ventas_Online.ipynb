##**Etapa 1: Abrir la base de datos con Pandas y aplicar json_normalize.**

---



```python
import pandas as pd
```

```python
import numpy as np
```

```python
datos_clientes = pd.read_json('dados_vendas_clientes.json')
datos_clientes.head()
```

```python
datos_clientes_norm = pd.json_normalize(datos_clientes['dados_vendas'])
```

```python
datos_clientes_norm.head()
```

##**Etapa 2**

```python
columnas = list(datos_clientes_norm.columns)
```

```python
columnas
```

```python
datos_clientes_norm = datos_clientes_norm.explode(columnas[1:])
```

```python
datos_clientes_norm.reset_index(drop=True, inplace=True)
```

```python
datos_clientes_norm
```

```python
datos_clientes_norm.info()
```

```python
datos_clientes_norm['Valor da compra']
```

```python
datos_clientes_norm['Valor da compra'] = datos_clientes_norm['Valor da compra'].apply(lambda x: x.replace('R$','').replace(',','.').strip())
```

```python
datos_clientes_norm.info()
```

```python
datos_clientes_norm['Valor da compra'] = datos_clientes_norm['Valor da compra'].astype(np.float64)
```

```python
datos_clientes_norm.info()
```

##**Etapa 3**

```python
datos_clientes_norm.head()
```

```python
datos_clientes_norm['Cliente'] = datos_clientes_norm['Cliente'].str.lower()
```

```python
datos_clientes_norm['Cliente'] = datos_clientes_norm['Cliente'].str.replace('[^a-zA-Z\-\']',' ', regex=True)
```

```python
datos_clientes_norm['Cliente'] = datos_clientes_norm['Cliente'].str.strip()
```

```python
datos_clientes_norm['Cliente']
```

##**Etapa 4**

```python
datos_clientes_norm.info()
```

```python
datos_clientes_norm['Data de venda'] = pd.to_datetime(datos_clientes_norm['Data de venda'], format='%d/%m/%Y')
```

```python
datos_clientes_norm
```

```python
datos_clientes_norm.info()
```

```python
total_compras = datos_clientes_norm.groupby(['Cliente'])['Valor da compra'].sum()
```
